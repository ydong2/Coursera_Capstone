```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius,
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['District', 
                  'District Latitude', 
                  'District Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    return(nearby_venues)
```

```python
sanmateo_venues = getNearbyVenues(names=sanmateo_df['District'],
                                   latitudes=sanmateo_df['Latitude'],
                                   longitudes=sanmateo_df['Longitude']
                                  )
```

```python
# Analyze each district
```

```python
# one hot encoding
sanmateo_onehot = pd.get_dummies(sanmateo_venues[['Venue Category']], prefix="", prefix_sep="")

# add district column back to dataframe
sanmateo_onehot['District'] = sanmateo_venues['District'] 

# move district column to the first column
fixed_columns = [sanmateo_onehot.columns[-1]] + list(sanmateo_onehot.columns[:-1])
sanmateo_onehot = sanmateo_onehot[fixed_columns]

sanmateo_onehot.head()
```

```python
sanmateo_onehot.shape
```

```python
sanmateo_grouped = sanmateo_onehot.groupby('District').mean().reset_index()
sanmateo_grouped
```

```python
sanmateo_grouped.shape
```

```python
#Sort the venues in descending order
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# Display top 10 venues for each district
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['District']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
district_venues_sorted = pd.DataFrame(columns=columns)
district_venues_sorted['District'] = sanmateo_grouped['District']

for ind in np.arange(sanmateo_grouped.shape[0]):
    district_venues_sorted.iloc[ind, 1:] = return_most_common_venues(sanmateo_grouped.iloc[ind, :], num_top_venues)

district_venues_sorted.head()
```

```python
#Cluster districts
```

```python
# set number of clusters
kclusters = 5

sanmateo_grouped_clustering = sanmateo_grouped.drop('District', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(sanmateo_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
district_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

sanmateo_merged = sanmateo_df

# merge sanmateo_grouped with sanmateo_df to add latitude/longitude for each district
sanmateo_merged = sanmateo_merged.join(district_venues_sorted.set_index('District'), on='District')

sanmateo_merged.head()
```

```python
sanmateo_merged=sanmateo_merged.drop(sanmateo_merged.index[14])
sanmateo_merged
```

```python
sanmateo_merged['Cluster Labels'] =sanmateo_merged['Cluster Labels'].astype(int)
sanmateo_merged[['Latitude', 'Longitude']] =sanmateo_merged[['Latitude', 'Longitude']].astype(float)
```

```python
#Visualize the resulting clusters
```

```python
# create map
map_clusters = folium.Map(location=[37.4337, -122.4014], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(sanmateo_merged['Latitude'], sanmateo_merged['Longitude'], sanmateo_merged['District'], sanmateo_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster],
        fill=True,
        fill_color=rainbow[cluster],
        fill_opacity=0.7).add_to(map_clusters)     
map_clusters
```
